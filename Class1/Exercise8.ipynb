```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#Ex1
img = np.array([[0, 0, 0.2, 0.5, 0.8, 1, 1],
       [0, 0, 0.2, 0.5, 0.8, 1, 1],
       [0, 0, 0.2, 0.5, 0.8, 1, 1],
       [0, 0, 0.2, 0.5, 0.8, 1, 1]]
)
mask = np.array([-1, 0, 1])

res = cv2.filter2D(img, -1, mask)

print(res)
```

```python
#Ex2
path = "imagens/xadrezado.jpg"

img = cv2.imread(path, cv2.COLOR_BGR2RGB)

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

mask1 = np.array([-2, 0, 2])
mask2 = np.array([-2, 0, 2]).reshape(1, -1)

res1 = cv2.filter2D(img, -1, mask1)
res2 = cv2.filter2D(img, -1, mask2)

cv2.imshow("img", res1)
cv2.imshow("img", res2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Ex3
path = "imagens/edificio.tif"

img = cv2.imread(path, cv2.COLOR_BGR2RGB)

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

mask1 = np.array([[1, 0, -1],
                 [1, 0, -1],
                 [1, 0, -1]])

mask2 = np.array([[1, 1, 1],
                 [0, 0, 0],
                 [-1, -1, -1]])

res1 = cv2.filter2D(img, -1, mask1)
res2 = cv2.filter2D(img, -1, mask2)
mag = np.sqrt(res1 ** 2 + res2 ** 2)

cv2.imshow("img", res1)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img", res2)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(mag, cmap='gray')
```

```python
#Ex4
path = "imagens/edificio.tif"

img = cv2.imread(path, cv2.COLOR_BGR2RGB)

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

mask1 = np.array([[1, 0, -1],
                  [2, 0, -2],
                  [1, 0, -1]])

mask2 = np.array([[1, 2, 1],
                  [0, 0, 0],
                  [-1, -2, -1]])

res1 = cv2.filter2D(img, -1, mask1)
res2 = cv2.filter2D(img, -1, mask2)
mag = np.sqrt(res1 ** 2 + res2 ** 2)

cv2.imshow("img", res1)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow("img", res2)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.imshow(mag, cmap='gray')
```

```python
#Ex5
path = "imagens/pulmao.jpg"

img = cv2.imread(path, cv2.COLOR_BGR2RGB)

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

img_canny = cv2.Canny(img, 100, 200)
img_lg = cv2.GaussianBlur(img, (5, 5), 0)
img_lg = cv2.Laplacian(img, cv2.CV_16S, 3)

plt.imshow(img_lg)

cv2.imshow("img", img_canny)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
