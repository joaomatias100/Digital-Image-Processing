```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#Ex1
path = "imagens/mamograma.tif"

plt.subplot(3, 1, 1)
img1 = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img1, cmap = "gray")

plt.subplot(3, 1, 2)
img2 = cv2.add(img1, 50)
plt.imshow(img2, cmap = "gray")

plt.subplot(3, 1, 3)
img3 = cv2.add(img1, 150)
plt.imshow(img3, cmap = "gray")
```

```python
#Ex2
path1 = "imagens/soma_1.jpg"
path2 = "imagens/soma_2.jpg"

img1 = cv2.imread(path1)
img2 = cv2.imread(path2)

print(img1.shape)
print(img2.shape)

if(img1.shape == img2.shape):
    img3 = cv2.add(img1, img2)
    plt.imshow(img3)
else:
    print("Erro")
```

```python
#Ex3
path1 = "imagens/dif1.bmp"
path2 = "imagens/dif2.bmp"

img1 = cv2.imread(path1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(path2, cv2.IMREAD_GRAYSCALE)

print(img1.shape)
print(img2.shape)

if(img1.shape[0] > img2.shape[0]):
    img1 = cv2.resize(img1, (img2.shape[0], img1.shape[1]))
else:
    img2 = cv2.resize(img2, (img1.shape[0], img2.shape[1]))
if(img1.shape[1] > img2.shape[1]):
    img1 = cv2.resize(img1, (img1.shape[0], img2.shape[1]))
else:
    img2 = cv2.resize(img2, (img2.shape[0], img1.shape[1]))

print(img1.shape)
print(img2.shape)

sub = cv2.subtract(img2, img1)
plt.imshow(sub)
  
```
