```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#Ex1
path = "imagens/mamograma.tif"

plt.subplot(3, 1, 1)
img1 = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img1, cmap = "gray")

plt.subplot(3, 1, 2)
img2 = cv2.add(img1, 50)
plt.imshow(img2, cmap = "gray")

plt.subplot(3, 1, 3)
img3 = cv2.add(img1, 150)
plt.imshow(img3, cmap = "gray")
```

```python
#Ex2
path1 = "imagens/soma_1.jpg"
path2 = "imagens/soma_2.jpg"

img1 = cv2.imread(path1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(path2, cv2.IMREAD_GRAYSCALE)

print(img1.shape)
print(img2.shape)

plt.subplot(3, 1, 1)
plt.imshow(img1, cmap = 'gray')
plt.subplot(3, 1, 2)
plt.imshow(img2, cmap = 'gray')

height1, width1 = img1.shape[0], img1.shape[1]
height2, width2 = img2.shape[0], img2.shape[1]

min_width = min(width1, width2)
min_height = min(height1, height2)

img1 = cv2.resize(img1, (min_height, min_width))
img2 = cv2.resize(img2, (min_height, min_width))

add = cv2.add(img1, img2)

plt.subplot(3, 1, 3)
plt.imshow(add, cmap = 'gray')
```

```python
#Ex3
path1 = "imagens/dif1.bmp"
path2 = "imagens/dif2.bmp"

img1 = cv2.imread(path1, cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread(path2, cv2.IMREAD_GRAYSCALE)

sub = cv2.subtract(img2, img1)

plt.subplot(2, 1, 1)
plt.imshow(sub, cmap = 'gray')

height, width = sub.shape[0], sub.shape[1]

final_sub = np.where(sub < 25, sub, 0)
plt.subplot(2, 1, 2)
plt.imshow(final_sub, cmap = 'gray')

 
```
