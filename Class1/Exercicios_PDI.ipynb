```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
#Ex1
path = "imagens/rx_mao.jpg"

img = cv2.imread(path)
plt.imshow(img)
plt.title("Rx Mao")
cv2.imwrite("Rx_gravado.tif", img)
```

```python
#Ex2
path1 = "imagens/rx_mao.jpg"
path2 = "imagens/rx_torax.jpg"
path3 = "imagens/Rx_bexiga.jpg"
path4 = "imagens/brain.jpg"

img1 = cv2.imread(path1)
img2 = cv2.imread(path2)
img3 = cv2.imread(path3)
img4 = cv2.imread(path4)


plt.subplot(2, 2, 1)
plt.imshow(img1)
plt.title("RX Mao")

plt.subplot(2, 2, 2)
plt.imshow(img2)
plt.title("RX Torax")

plt.subplot(2, 2, 3)
plt.imshow(img3)
plt.title("RX Bexiga")

plt.subplot(2, 2, 4)
plt.imshow(img4)
plt.title("Brain")
```

```python
#Ex4
path = "imagens/retina.tif"

img = cv2.imread(path)

red, green, blue = cv2.split(img)
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.subplot(2, 2, 2)
plt.imshow(red)
plt.subplot(2, 2, 3)
plt.imshow(green)
plt.subplot(2, 2, 4)
plt.imshow(blue)
```

```python
#Ex5
img = np.ones((255, 255, 3))

x, y = 90, 70
height = 100
width = 100

x2 = x + width
y2 = y - height

cv2.rectangle(img, (x2, y2), (x, y), (0, 0, 0), thickness = -1)

plt.imshow(img)
```
